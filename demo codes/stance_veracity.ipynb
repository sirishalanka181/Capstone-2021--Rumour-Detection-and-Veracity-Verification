```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Dropout, LSTM
from keras.layers import TimeDistributed, Masking
from keras import optimizers
from keras import regularizers
from keras.models import load_model
import json
import os
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import hamming_loss
from sklearn.metrics import multilabel_confusion_matrix



def predict_model_veracity(x_test, y_test):
    model=load_model('/content/drive/MyDrive/CAPSTONE PROJECT - PRAJNA SIRISHA SUKANYA/code/Stance_Veracity/my_model_veracity.h5')
    Y_pred = model.predict_classes(x_test)
    labels=[]
    true_labels=[]
    print(Y_pred[:10])
    # for i in range(len(x_test)):
    #     print("X=%s, Predicted=%s" % (x_test[i], Y_pred[i]))
    
    print(classification_report(y_test,Y_pred))
    Y_pred=np.ndarray.tolist(Y_pred)
    for i in range(len(Y_pred)):
        if(Y_pred[i]==0):
            labels.append("true")
        elif(Y_pred[i]==1):
            labels.append("false")
        else:
            labels.append("unverified")

    for i in range(len(y_test)):
        if(y_test[i]==0):
            true_labels.append("true")
        elif(y_test[i]==1):
            true_labels.append("false")
        else:
           true_labels.append("unverified")
    # print(labels)
    # print(true_labels)
    print("predicted","true label")
    for i in range(len(y_test)):
        print(labels[i],true_labels[i],end="\n")

def predict_model_stance(x_test,y_test):
    model=load_model('/content/drive/MyDrive/CAPSTONE PROJECT - PRAJNA SIRISHA SUKANYA/code/Stance_Veracity/my_model_stance.h5')
    Y_pred=model.predict_classes(x_test)
    y_test=np.ndarray.tolist(y_test)
    Y_pred= np.ndarray.tolist(Y_pred)
    labels=[]
    true_labels=[]
    correct=0
    for i in range(len(y_test)):
        if(int(y_test[i][1])==Y_pred[i][1]):
            correct+=1
        if(Y_pred[i][1]==0):
            labels.append("support")
        elif(Y_pred[i][1]==1):
            labels.append("deny")
        elif(Y_pred[i][1]==2):
            labels.append("query")
        else:
            labels.append("comment")

    for i in range(len(y_test)):
        
        if(y_test[i][1]==0):
            true_labels.append("support")
        elif(y_test[i][1]==1):
            true_labels.append("deny")
        elif(y_test[i][1]==2):
            true_labels.append("query")
        else:
            true_labels.append("comment")

    print("predicted","true label")
    for i in range(len(y_test)):
        print(labels[i],true_labels[i],end="\n")
    print("Correct predictions",correct)
    print("Total number if tweets considered", len(y_test))
    print("Accuracy is ", correct/len(y_test))
    # print(labels)
    # print(true_labels)

    

x_test_ver = np.load('/content/drive/MyDrive/CAPSTONE PROJECT - PRAJNA SIRISHA SUKANYA/code/Stance_Veracity/test/train_array.npy')
y_test_ver = np.load('/content/drive/MyDrive/CAPSTONE PROJECT - PRAJNA SIRISHA SUKANYA/code/Stance_Veracity/test/labels.npy')
# print(len(x_test_ver))
# print(y_test_ver[:10])

x_test_stance = np.load('/content/drive/MyDrive/CAPSTONE PROJECT - PRAJNA SIRISHA SUKANYA/code/Stance_Veracity/test/train_array.npy')
y_test_stance = np.load('/content/drive/MyDrive/CAPSTONE PROJECT - PRAJNA SIRISHA SUKANYA/code/Stance_Veracity/test/fold_stance_labels.npy')
print(len(x_test_stance))
# print(y_test_stance[:10])
predict_model_stance(x_test_stance,y_test_stance)
predict_model_veracity(x_test_ver,y_test_ver)

```
