```python
import pickle
from sklearn.feature_extraction import DictVectorizer
import json

file='/content/drive/MyDrive/CAPSTONE PROJECT - PRAJNA SIRISHA SUKANYA/code/tweet_structure.json'
file=open(file)
tweet_dict=json.load(file)
tweets=[]
tweets.append(tweet_dict)
print(tweets)
v = DictVectorizer(sparse=False)
tweets_x=v.fit_transform(tweets)

filename = '/content/drive/MyDrive/CAPSTONE PROJECT - PRAJNA SIRISHA SUKANYA/code/dec_tree.pkl'
loaded_model = pickle.load(open(filename, 'rb'))

y_pred=loaded_model.predict(tweets_x)

for pred in y_pred:
  if(pred==0):
    print("Non-Rumour")
  else:
    print("Rumour")
```
